# Computational Challenge 8 Group 2

## Problem 1

With the given payoff matrix we see that the expected outcome of strategy $p$ given that opponent uses strategy $q$ is 

$$E[p]=\frac{B-C}{2}\cdot p\cdot q+B\cdot(1-q)\cdot p+\frac{B}{2}\cdot(1-p)\cdot(1-q) $$ 

the expeted payoff for $q$ is the same formula as above, but the $q$ and $p$ are interchanged. We are assuming that $B,C>0$  

```python
import numpy as np
import random
from scipy.special import gamma, factorial
from scipy import integrate
import matplotlib.pyplot as plt
import seaborn as sns
import collections
import tqdm
from numba import jit
from numba import njit, prange
import time
from scipy.integrate import solve_ivp
from matplotlib.pyplot import cm
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
B=2
C=1
p=[]
q=[]
x=[]
Ep=0
Average=[]
mx=[0]*101
method=[0]*101
for j in range(101):
    q=np.append(q,j/100)
    x=np.append(x,j/100)
    Average=np.append(Average,0)
    for i in range(101):
        p=np.append(p,i/100)
        Ep=((B-C)/2*p[i]*q[j]+B*(1-q[j])*p[i]+B/2*(1-p[i])*(1-q[j]))
        if mx[j]<Ep:
            mx[j]=Ep
            method[j]=i/100
print('B=',B)
print('C=', C)
plt.scatter(method,x)
plt.title("Optimal p strategy with respect to all strategies q")
plt.xlabel('Optimal Strategy for p')
plt.ylabel('Given Strategy for q')
plt.show()
plt.scatter(x,mx)
plt.title('Expected Payoff for optimal p')
plt.xlabel('Given q strategy')
plt.ylabel('Expected Payoff')
```

## Problem 2

Given that there is a population of $N$ with $i$ hawks and $N-i$ doves. We see that the fitnesses are 

$$f_{hawk}=1-\omega +\omega\frac{(i-1)(B-C)/2+(N-i)B}{N-1}$$

$$f_{dove}=1-\omega+\omega\frac{(N-i-1)B/2}{N-1}$$

where $\omega$ is the intensity of selection. 

## Problem 3 



```python
N=1000
B=4
C=6
mu=.001
w=1
f_hawk=0
f_dove=0
hawks=10
graph_hawk=[]
graph_dove=[]
for j in range(300):
    hawks=10
    for i in range(20000):
        if random.random() < mu:
            if random.random() < (N-hawks)/N:
                hawks=hawks+1
            else:
                hawks=hawks-1
            if hawks<0:
                hawks=0
            if hawks>1000:
                hawks=1000
        f_hawks=1-w+w*((hawks-1)*(B-C)/2+(N-hawks)*B)/(N-1)
        f_doves=1-w+w*((N-hawks-1)*B/2)/(N-1)
        p_hawks=(hawks*f_hawks)/(hawks*f_hawks+(N-hawks)*f_doves)*(N-hawks)/N
        p_doves=((N-hawks)*f_doves)/(hawks*f_hawks+(N-hawks)*f_doves)*(hawks)/N
        r=random.random()
        if r<p_hawks:
            hawks=hawks+1
        elif r<p_hawks+p_doves:
            hawks=hawks-1
    graph_hawk=np.append(graph_hawk,hawks)
    graph_dove=np.append(graph_dove,N-hawks)
plt.hist(graph_hawk,bins=20,label='hawk')
plt.hist(graph_dove,bins=20,label='dove')
plt.title("Hawks vs Doves B=4")
plt.xlabel('Final total in simulation')
plt.ylabel('Total amount in each bin')
plt.legend(loc='upper center')
plt.show()
```
